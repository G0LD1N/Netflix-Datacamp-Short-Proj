1. What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.

For the 'frequent' duration meaning 'common' or 'many times' the duration i'll use "mode" but for that i'll need to import a python library called 'scipy'.

```python
from scipy import stats

# Filter the df for movies released in the 1990s.
movies_1990s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000) & (netflix_df['type'] == 'Movie')]

# Find the mode of the 'duration' column
mode_result = stats.mode(movies_1990s['duration'])
#duration = int(mode_result.mode[0])

print(mode_result) # 94
```

2. A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
# Filter the df for short action movies released in the 1990s.
short_movies_1990s = netflix_df[
    (netflix_df['duration'] < 90) &
    (netflix_df['genre'] == 'Action') & 
    (netflix_df['release_year'] >= 1990) & 
    (netflix_df['release_year'] < 2000) & 
    (netflix_df['type'] == 'Movie')
]

# Count the number of short action movies released in the 1990s.
short_movie_count = short_movies_1990s.shape[0]
print(short_movie_count)
```
